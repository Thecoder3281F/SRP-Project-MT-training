```python
import csv

lst = ["test", "train", "val"]


for element in lst:
    # Input text files
    file1 = f"../MIT/MIT_separated/src-{element}.txt"
    file2 = f"../MIT/MIT_separated/tgt-{element}.txt"

    # Output CSV file
    output = f"../MIT/MIT_separated/MIT_separated_{element}.csv"

    # Read both text files
    with open(file1, 'r', encoding='utf-8') as f1, open(file2, 'r', encoding='utf-8') as f2:
        lines1 = [line.strip() for line in f1]
        lines2 = [line.strip() for line in f2]

    # Pad shorter list so lengths match
    max_len = max(len(lines1), len(lines2))
    lines1 += [''] * (max_len - len(lines1))
    lines2 += [''] * (max_len - len(lines2))

    # Write to CSV
    with open(output, 'w', newline='', encoding='utf-8') as csvfile:
        writer = csv.writer(csvfile)
        writer.writerow(['input', 'target'])
        writer.writerows(zip(lines1, lines2))

    print(f"Merged {file1} and {file2} into {output}")
```
