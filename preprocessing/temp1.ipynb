```python
import pandas as pd
import os
import glob

# Get all CSV files in the processed folder
csv_files = glob.glob('../processed/*.csv')

for file_path in csv_files:
    # Read the CSV file
    df = pd.read_csv(file_path)
    
    base_name = os.path.splitext(os.path.basename(file_path))[0]
    
    # Create separate reactants + products file
    separate_df = pd.DataFrame({
        'input_reaction': df['input_reaction_separated'],
        'product_1_tokenized': df['product_1_tokenized'],
        'product_2_tokenized': df['product_2_tokenized']
    })
    separate_df.to_csv(f'../processed/{base_name}_separate.csv', index=False)
    
    # Create combined reactants + products file
    # Combine reactants with '+'
    combined_df = pd.DataFrame({
        'input_reaction': df['input_reaction_combined'],
        'product_1_tokenized': df['product_1_tokenized'],
        'product_2_tokenized': df['product_2_tokenized']
    })
    combined_df.to_csv(f'../processed/{base_name}_combined.csv', index=False)

print("Files have been split successfully!")
```
